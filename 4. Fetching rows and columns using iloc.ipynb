```python
import pandas as pd
```

```python
data = pd.read_csv("matches.csv")
```

```python
data['winner'] # fetched a column ans hence its a series
```

```python
type(data['winner']) # pandas.core.series.Series
```

1. Dataframe - rows and columns
2. Series - index and value

```python
data[['team1', 'team2', 'winner']] # fetching multiple columns
```

```python
type(data[['team1', 'team2', 'winner']]) # pandas.core.frame.DataFrame
```

```python
# fetching rows
data.iloc[0] # series
```

```python
# indexing
data.iloc[1:3]
```

```python
type(data.iloc[1:3]) # pandas.core.frame.DataFrame
```

```python
data.iloc[1:11:2] # similar to python normal indexing, slicing and steps
```

```python
# Fancy indexing similar to in Numpy
data.iloc[[1,5,8]]
```

```python
# fetching columns using 'iloc'
data.iloc[:, [4,5,10]] # data.iloc[fetching rows, fetching columns]
```
