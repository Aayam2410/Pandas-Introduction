```python
import pandas as pd
```

```python
data = pd.read_csv('matches.csv')
```

When we work in Machine Learning if we send duplicate data to our model it will only decrease the efficiency of our model and leads to poor performance.

```python
# not inplace by default, to make inplace you can pass 'inplace=True'
data.drop_duplicates(subset=['city'])
```

```python
data.drop_duplicates(subset=['city']).shape
```

```python
# 'keep' is a keyword argument which provides us an option to save the 'first' or 'last' occurence of our particular data
# by default the 'keep' argument is equal to 'first'

data.drop_duplicates(subset=['city'], keep='last')
```

```python
# we can apply drop_duplicates() on multiple columns

data.drop_duplicates(subset=['city', 'season']).shape
```

```python
# show the winner of each season alongwith the year

data.drop_duplicates('season', keep='last').sort_values('season', ascending=True)[['season', 'winner']]
```
