To understand the working of isin() fucntion we'll go through a question and try to find its solution.

Question -> Find the most destructive death over batsman in the history of IPL.

Conditions:-
1. We'll calculate the solution on the basis of strike rate of a batsman.
Strike rate = (Number of runs/Number of balls)*100

2. Minimum balls played by the batsman in between the over 16th - 20th should be 200.

```python
import pandas as pd
```

```python
delivery = pd.read_csv('deliveries.csv')
```

```python
delivery.shape
```

```python
# fetching only overs from 16-20 (both included)
mask = delivery['over']>15 # boolean series
death_over = delivery[mask] 
```

```python
# Fetches all the players and no. of balls played by them from 16-20 over

all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
all_batsman
```

```python
# applying one of our conditions i.e. played 200 balls or more

x = all_batsman > 200
x
```

```python
# now applying masking on "all_batsman" with 'x'

y = all_batsman[x]
y
```

```python
# now we have to extract the indices of 'y'

y.index
```

```python
# index object to list

batsman_list = y.index.tolist()
batsman_list
```

```python
# runs scored by all these 43 batsmen
# balls played by all these 43 batsmen
# here we can't make 43 mask variables and thus isin() function comes into play

mask = delivery['batsman'].isin(batsman_list) # boolean list
final = death_over[death_over['batsman'].isin(batsman_list)]
final
```

```python
# calculating runs

runs = final.groupby('batsman')['batsman_runs'].sum()
runs
```

```python
# calculating balls

balls = final.groupby('batsman')['batsman_runs'].count()
balls
```

```python
sr = (runs/balls) * 100
sr.sort_values(ascending=False)
```
