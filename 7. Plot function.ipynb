```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('matches.csv')
```

```python
# during data analysis, data visualization is ver important because it helps us to look at patterns via plots, graphs, pie charts, etc.

data['winner'].value_counts().plot(kind='bar')
```

```python
# top 5 teams only, winner --> categorical datatype

data['winner'].value_counts().head().plot(kind='bar')
```

```python
# horizontal bar
data['winner'].value_counts().plot(kind='barh')
```

```python
# pie chart, toss_decision --> categorical datatype
data['toss_decision'].value_counts().plot(kind='pie')
```

```python
# histogram, win_by_runs --> numerical datatype

data['win_by_runs'].plot(kind='hist')
```
